```r
rm(list=ls())
library(ggplot2)
```

```r
data=read.table("gwas.summary",header=TRUE,sep='\t')
```

```r
exposure_names=c("X12am_6am","UsualWalkingPace","DurationWalkingForPleasure","NumberDaysWalked10Minutes")
outcome_names=c("DeathAge","CvdStatus")
```

## Exposure beta vs outcome p

```r
for(exposure_name in exposure_names){
    for(outcome_name in outcome_names){
exposure_subset=data[data$Trait==exposure_name,]
exposure_subset=exposure_subset[order(exposure_subset$SNP),]
snps=exposure_subset$SNP 

outcome_subset= data[(data$Trait==outcome_name) & (data$SNP %in% snps),]
outcome_subset=outcome_subset[order(outcome_subset$SNP),]

exposure_beta=exposure_subset$BETA
exposure_se=exposure_subset$SE 
exposure_p=exposure_subset$PVAL
outcome_beta=outcome_subset$BETA 
outcome_se = outcome_subset$SE
outcome_p=outcome_subset$PVAL
        
toplot=data.frame(exposure_beta,outcome_beta,outcome_p)
names(toplot)=c("exposure_beta","outcome_beta","outcome_p")
p=ggplot(data=toplot)+
    geom_point(aes(x=exposure_beta,y=outcome_p))+
        ggtitle(paste(exposure_name,'vs',outcome_name,sep=' '))
        print(p)
    }
} 
```
